```python
from collaborativefilter.model import CollaborativeFilter

from utilities import data

import numpy as np 
np.set_printoptions(suppress=True)

path = './utilities/ratings.csv'
```

```python
#data prep
data_ = data.load(path)
data_ = data.transform(data_)

train,test = data.split(data_)

user_ids = list(test.userId)
movie_ids = list(test.movieId)
test_ratings = list(test.rating)
```

```python
#instantiate most recent trained model 
cf = CollaborativeFilter()
trained_model = cf.load_trained_model()
```

```python
#predicting 10 ratings for the first user in test sample using trained model 
#note that only 5 epochs were used
ratings=list()
user_id = np.array([user_ids[0]])

for i in range(10):
    movie_id = np.array([movie_ids[i]])
    prediction = trained_model.predict([user_id,movie_id])[0][0]
    ratings.append('actual: {},prediction: {}'.format(test_ratings[i],prediction))
print(ratings)
#looks good!
```

```python
#calculating mean absolute error over entire test sample
from sklearn.metrics import mean_absolute_error

predictions = trained_model.predict([test.userId,test.movieId])
mae = mean_absolute_error(test_ratings,predictions)

print('mean absolute error of test sample: {}'.format(mae))
```

```python
#an interesting diversion is to peek at the user embedding layer and visualize how 
#users cluster together 

from sklearn.metrics.pairwise import euclidean_distances
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt 

def embeddings(cf,layer,labels):
    
    embedding = cf.get_weights()[layer]
    embedding = embedding[2:]
    
    labels = list(labels)
    
    distance_matrix = euclidean_distances(embedding)
    
    tsne = TSNE(n_components=2,random_state=0,n_iter=10000,perplexity=10)
    
    T = tsne.fit_transform(distance_matrix)
    
    plt.figure(figsize=(20,10))
    plt.scatter(T[:,0],T[:,1])
    
    for label,x,y in zip(labels,T[:,0],T[:,1]):
        plt.annotate(label,xy=(x,y))

    plt.show()

embeddings(trained_model,0,user_ids)
```
